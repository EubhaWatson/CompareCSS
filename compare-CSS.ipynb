```python
import os
import difflib

def compare_html_files(dir):
    results = []

    # Walk through all files in the directory and its subdirectories
    for root, dirs, files in os.walk(dir):
        for file in files:
            # Check if the file is an HTML file and exclude the reset code
            if file.endswith(".html"):
                # Read the contents of the CSS file
                file_path = os.path.join(root, file)
                with open(file_path, "r") as f:
                    html1 = f.read()

                # Compare the file to all other HTML files
                for root2, dirs2, files2 in os.walk(dir):
                    for file2 in files2:
                        if file2.endswith(".html"):
                            file_path2 = os.path.join(root2, file2)
                            with open(file_path2, "r") as f2:
                                html2 = f2.read()

                            # Calculate the percentage similarity
                            if html1 != html2:
                                s = difflib.SequenceMatcher(None, html1, html2)
                                percentage = round(s.ratio() * 100, 2)

                                # Add the result to the list
                                results.append((file_path, file_path2, percentage))

    # Sort the results by percentage similarity
    results.sort(key=lambda x: x[2], reverse=True)

    # Output the results to a text file
    with open("/Users/watson/Desktop/results.txt", "w") as f:
        for result in results:
            f.write(f"{result[0]} and {result[1]}: {result[2]}%\n")
            
    #Let the user know it's done.
    print("Your file is written and ready.")

# Main Input Loop
check = 0

while (check == 0):
    # Prompt the user for a directory
    dir = input("Enter the directory path: ")

    # Make sure the directory exists
    if not os.path.isdir(dir):
        print(f"Error: {dir} is not a valid directory.")
    else:
        check = 1
        # Try to compare the files
        try:
            compare_html_files(dir)
        except Exception as e:
            print(f"Error: {e}")
```

```python
# This variation includes the actual diff output (i.e. insertions and deletions).

import os
import difflib
from itertools import combinations

def get_similarity(file1_lines, file2_lines):
    matcher = difflib.SequenceMatcher(None, file1_lines, file2_lines)
    return matcher.ratio()

def compare_files(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        file1_lines = f1.readlines()
        file2_lines = f2.readlines()

    similarity = get_similarity(file1_lines, file2_lines)
    return file1, file2, similarity, difflib.unified_diff(file1_lines, file2_lines, fromfile=file1, tofile=file2, lineterm='')

def find_css_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.css') and file not in ['reset.css', 'modern-reset.css']:
                yield os.path.join(root, file)

def compare_all_css_files(directory):
    css_files = list(find_css_files(directory))
    # Debug: Print out the CSS files found
    print(f"CSS files to compare: {css_files}")

    if not css_files or len(css_files) < 2:
        print("Not enough CSS files to compare.")
        return
    
    results = []
    for file1, file2 in combinations(css_files, 2):
        file1, file2, similarity, diff = compare_files(file1, file2)
        results.append((file1, file2, similarity, '\n'.join(diff)))

    # Sort by similarity in descending order
    results.sort(key=lambda x: x[2], reverse=True)

    output_file_path = '/Users/watson/Desktop/css-results.txt'
    with open(output_file_path, 'w') as output_file:
        for file1, file2, similarity, diff in results:
            output_file.write(f"Differences between {file1} and {file2}:\n")
            output_file.write(f"Similarity: {similarity*100:.2f}%\n")
            output_file.write(diff)
            output_file.write("\n\n")

    print(f"Results written to {output_file_path}")

# Prompting the user to input the directory path
directory_path = input("Please enter the path to the directory containing CSS files: ")
compare_all_css_files(directory_path)
```
